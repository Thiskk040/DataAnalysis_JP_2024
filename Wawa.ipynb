```python
import pandas as pd

# Load the CSV file
file_path = 'Merged/0004_Man.csv'
data = pd.read_csv(file_path)

# Extract rows where Character_status is empty
initial_character_status = data[data['Character_status'].isna()]

# Select the first row where User_Action is 'Answer'
initial_answer = initial_character_status[initial_character_status['User_Action'] == 'Answer'].iloc[0]

initial_result = {
    'Character_position.x': initial_answer['Character_position.x'],
    'time_rounded': initial_answer['time'],
    'Form_Question': initial_answer['Form_Question'],
    'Form_Answer': initial_answer['Form_Answer'],
    'HRV_LF/HF': initial_answer['HRV_LFHF']
}

# Extract the remaining rows based on Character_status
remaining_data = data.dropna(subset=['Character_status'])
remaining_results = []

# Iterate over unique Character_status values
for status in remaining_data['Character_status'].unique():
    # Select the first row where User_Action is 'Answer' for each status
    status_data = remaining_data[(remaining_data['Character_status'] == status) & 
                                 (remaining_data['User_Action'] == 'Answer')]
    if not status_data.empty:
        first_answer = status_data.iloc[0]
        result = {
            'Character_status': status,
            'time_rounded': first_answer['time'],
            'Form_Question': first_answer['Form_Question'],
            'Form_Answer': first_answer['Form_Answer'],
            'HRV_LF/HF': first_answer['HRV_LFHF']
        }
        remaining_results.append(result)

# Calculate mean HRV_LF/HF for each Character_status
mean_hrv_lfhf = remaining_data.groupby('Character_status')['HRV_LFHF'].mean().reset_index()

# Convert initial_result to DataFrame
initial_df = pd.DataFrame([initial_result])

# Convert remaining_results to DataFrame
remaining_df = pd.DataFrame(remaining_results)

# Add a new column to differentiate data
initial_df['Data_Type'] = 'Initial'
remaining_df['Data_Type'] = 'Remaining'
mean_hrv_lfhf['Data_Type'] = 'Mean'

# Combine all DataFrames
combined_df = pd.concat([initial_df, remaining_df, mean_hrv_lfhf], ignore_index=True)

# Save combined data to a single CSV file
combined_df.to_csv('combined_results.csv', index=False)
```
